### Perform tokenization (Whitespace, Punctuation-based, Treebank, Tweet, MWE) using NLTK library. Use porter stemmer and snowball stemmer for stemming. Use any technique for lemmatization. 

```python
import nltk

from nltk.tokenize import (
    WhitespaceTokenizer,
    wordpunct_tokenize,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer
)

```

```python
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "NLTK is awesome! I'm learning NLP, AI & Machine Learning :) #Excited"
print(text)
```

```python
wt = WhitespaceTokenizer()
whitespace_tokens = wt.tokenize(text)

print("Whitespace Tokenization:")
print(whitespace_tokens)
```

```python
punct_tokens = wordpunct_tokenize(text)

print("\nPunctuation-Based Tokenization:")
print(punct_tokens)
```

```python
treebank = TreebankWordTokenizer()
treebank_tokens = treebank.tokenize(text)

print("\nTreebank Tokenization:")
print(treebank_tokens)
```

```python
tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(text)

print("\nTweet Tokenization:")
print(tweet_tokens)
```

```python
mwe = MWETokenizer([('Machine', 'Learning'), ('Natural', 'Language', 'Processing')])
mwe_tokens = mwe.tokenize(text.split())

print("\nMWE Tokenization:")
print(mwe_tokens)
```

```python
words = ["running", "runs", "ran", "easily", "fairly"]
print(words)
```

```python
porter = PorterStemmer()
porter_stems = [porter.stem(word) for word in words]

print("\nPorter Stemmer:")
print(porter_stems)
```

```python
snowball = SnowballStemmer("english")
snowball_stems = [snowball.stem(word) for word in words]

print("\nSnowball Stemmer:")
print(snowball_stems)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

print("\nLemmatization:")
print(lemmatized_words)
```

```python
lemmatized_verbs = [lemmatizer.lemmatize(word, pos='v') for word in words]
print("\nLemmatization with POS:")
print(lemmatized_verbs)
```
