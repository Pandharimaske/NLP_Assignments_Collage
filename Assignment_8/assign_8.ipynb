```python
import spacy
from sklearn.metrics import classification_report, accuracy_score
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
texts = [
    "Elon Musk founded SpaceX in the United States.",
    "Apple announced the new iPhone in California.",
    "Barack Obama was the president of the USA.",
    "Google headquarters is located in Mountain View."
]

texts
```

```python
true_entities = [
    ["PERSON","ORG","GPE"],
    ["ORG","PRODUCT","GPE"],
    ["PERSON","GPE"],
    ["ORG","GPE"]
]
```

```python
predicted_entities = []

for text in texts:
    doc = nlp(text)
    predicted_entities.append([ent.label_ for ent in doc.ents])
    
```

```python
y_true = []
y_pred = []

for t, p in zip(true_entities, predicted_entities):
    min_len = min(len(t), len(p))
    
    y_true.extend(t[:min_len])
    y_pred.extend(p[:min_len])
```

```python
print("Accuracy:", accuracy_score(y_true, y_pred))

print(classification_report(y_true, y_pred))
```

```python
for text in texts:
    doc = nlp(text)
    
    print("\nText:", text)
    
    for ent in doc.ents:
        print(ent.text, "->", ent.label_)
```
