### Use WordNet to identify semantic relationships such as synonymy, antonymy, and hypernymy in text data

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize
import pandas as pd
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = """
The car is very fast and good. 
A happy person usually drives a vehicle carefully. 
Sometimes a bad driver causes accidents.
"""

print(text)
```

```python
tokens = word_tokenize(text)
tokens
```

```python
def get_wordnet_relations(word):
    
    synonyms = set()
    antonyms = set()
    hypernyms = set()
    
    synsets = wn.synsets(word)
    
    for syn in synsets:
        
        # synonyms
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
            
            # antonyms
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())
        
        # hypernyms
        for hyper in syn.hypernyms():
            hypernyms.add(hyper.name().split('.')[0])
    
    return list(synonyms), list(antonyms), list(hypernyms)
```

```python
results = []

for word in tokens:
    
    syn, ant, hyp = get_wordnet_relations(word.lower())
    
    if syn or ant or hyp:
        results.append({
            "Word": word,
            "Synonyms": syn,
            "Antonyms": ant,
            "Hypernyms": hyp
        })
```

```python
df = pd.DataFrame(results)
df
```

```python
for index, row in df.iterrows():
    
    print("Word:", row["Word"])
    print("Synonyms:", row["Synonyms"])
    print("Antonyms:", row["Antonyms"])
    print("Hypernyms:", row["Hypernyms"])
    print("-"*40)
```

```python
df.to_csv("wordnet_relationships.csv", index=False)
```

```python
word = "car"

syn, ant, hyp = get_wordnet_relations(word)

print("Word:", word)
print("Synonyms:", syn)
print("Antonyms:", ant)
print("Hypernyms:", hyp)
```
